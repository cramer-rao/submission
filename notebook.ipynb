## Exploratory Data Analysis (EDA)

### Explore Pertanyaan 1

Untuk menjawab pertanyaan "Kategori produk apa yang paling banyak dan paling sedikit terjual?", pertama-tama akan dilakukan pembuatan DataFrame baru all_df dengan menggabungkan customers_df, order_items_df, products_df, dan orders_df dengan function merge(). Akan dilihat rangkuman parameter statistik dari all_df menggunakan method describe()

```python
sales_product_df = pd.merge(
  left=order_items_df,
  right=products_df,
  how="left",
  left_on="product_id",
  right_on="product_id"
)
orders_customers_df = pd.merge(
  left=orders_df,
  right=customers_df,
  how="left",
  left_on="customer_id",
  right_on="customer_id"
)

all_df = pd.merge(
  left=sales_product_df,
	right=orders_customers_df,
	how="left",
	left_on="order_id",
	right_on="order_id"
)

all_df["total_price"] = all_df["price"] * all_df["order_item_id"]
all_df["total_cost"] = (all_df["price"] + all_df["freight_value"]) * all_df["order_item_id"]
all_df["order_date"] = all_df["order_purchase_timestamp"].dt.date
all_df['order_date'] = pd.to_datetime(all_df['order_date'])
```

```python
all_df.describe()
```

Akan digunakan method groupby() untuk menghasilkan Pivot Table jumlah penjualan dari setiap kategori produk

```python
all_df.groupby("product_category_name").order_item_id.sum().sort_values(ascending=False)
```

Berikut adalah top 10 kategori produk dengan penjualan tertinggi

```python
top_selling_categories_df = all_df.groupby("product_category_name").order_item_id.sum().sort_values(ascending=False).reset_index().head(10)
print(top_selling_categories_df)
```

Berikut adalah top 10 kategori produk dengan penjualan terendah

```python
bottom_selling_categories_df = all_df.groupby("product_category_name").order_item_id.sum().sort_values().reset_index().head(10)
print(bottom_selling_categories_df)
```

**Insight:**
- Rata-rata harga produk yang dijual di e-commerce tersebut adalah 120.09 dengan harga terendah 0.85 dan harga tertinggi 6735. Terdapat variasi yang cukup besar dalam harga produk, ditunjukkan dengan nilai standar deviasi 182.04
- Kategori produk terlaris adalah "cama_mesa_banho" sebanyak 13484 items
- Kategori produk dengan penjualan paling sedikit adalah "seguros_e_servicos" sebanyak 2 items

### Explore Pertanyaan 3

Untuk menjawab pertanyaan "Bagaimana tren penjualan perusahaan dari waktu ke waktu?", akan dilihat data time_series bulanan yang memuat informasi jumlah transaksi per bulan serta revenue yang dihasilkan.

```python
monthly_orders_df = all_df.resample(rule='M', on='order_date').agg({
    "order_id": "nunique",
    "total_price": "sum"
})
monthly_orders_df.index = monthly_orders_df.index.strftime('%Y-%m')
monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    "order_id": "order_count",
    "total_price": "revenue"
}, inplace=True)

print(monthly_orders_df)
```

## Analisis Lanjutan (Opsional)

Akan dilakukan analisis lanjutan, yakni RFM Analysis dengan tujuan  mengelompokkan pelanggan berdasarkan perilaku pembelian mereka dengan memperhatikan tiga faktor utama: recency, frequency, dan monetary.

Tiga faktor tersebut dapat direpresentasikan dalam tiga pertanyaan berikut.
- Recency: Kapan terakhir pelanggan melakukan transaksi?
- Frequency: Seberapa sering seorang pelanggan melakukan pembelian dalam beberapa bulan terakhir?
- Monetary: Berapa banyak uang yang dihabiskan pelanggan dalam beberapa bulan terakhir?


Terlebih dahulu akan dibuat DataFrame yang menyimpan informasi terkait recency, frequency, dan monetary. Kemudian, dibuat visualisasi dari 5 best customer berdasarkan 3 parameter RFM.

```python
rfm_df = all_df.groupby(by="customer_unique_id", as_index=False).agg({
"order_purchase_timestamp": "max",
"order_id": "nunique",
"total_cost": "sum"
})
rfm_df.columns = ["customer_unique_id", "max_order_timestamp", "frequency", "monetary"]

rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date
recent_date = orders_df["order_purchase_timestamp"].dt.date.max()
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)

rfm_df.drop("max_order_timestamp", axis=1, inplace=True)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))
colors = ["#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4"]

sns.barplot(y="recency", x="customer_unique_id", data=rfm_df.sort_values(by="recency", ascending=True).head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency (days)", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=15, rotation=90)

sns.barplot(y="frequency", x="customer_unique_id", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15, rotation=90)

sns.barplot(y="monetary", x="customer_unique_id", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15, rotation=90)

plt.suptitle("Best Customer Based on RFM Parameters (customer_unique_id)", fontsize=20)
plt.show()
```

Pada visualisasi di atas, telah ditampilkan masing-masing 5 best customer berdasarkan setiap rfm parameters.
- Recency: 5 customer dengan jarak terdekat (0 hari) dari tanggal order terakhir dengan hari terakhir pada time series data yang digunakan untuk RFM Analysis
- Frequency: 5 customer dengan frekuensi transaksi terbanyak
- Monetary: 5 customer dengan jumlah pengeluaran tertinggi, mencakup total harga beserta ongkos kirim
